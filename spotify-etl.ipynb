```python
#reading dataframe
df = pd.read_csv('cleaned_dataset.csv')
df.head()
```

```python
# dataset description
df.info()
```

```python
#rename columns
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
#checking unique values in each columns
for col in df.columns:
    print(f"Unique values in '{col}':\n", df[col].unique(), "\n")
```

```python
#checking unique values in each columns
for col in df.columns:
    print(f"Number of values in '{col}':\n", df[col].nunique(), "\n")
```

```python
df.describe()
```

```python
#removing rows which have 0 as minimum duration_min - since these are synonymous to NaNs
df = df[df['duration_min'] != 0]
```

```python
#checking for null values
df.isna().sum()
```

```python
df.dropna(subset=['energyliveness'],inplace=True)
```

```python
df.licensed.value_counts()
df.official_video.value_counts()
```

```python
#converting to boolean columns
df['licensed'] = df['licensed'].astype(str).str.strip().str.lower().map({
    'true': 1,
    'false': 0,
    '1': 1,
    '0': 0
})

df['official_video'] = df['official_video'].astype(str).str.strip().str.lower().map({
    'true': 1,
    'false': 0,
    '1': 1,
    '0': 0
})
```

```python
df.licensed.value_counts(dropna=False)
df.official_video.value_counts(dropna=False)
```

```python
#resetting index
df.reset_index(inplace=True)
```

```python
df.rename(columns = {'index' : 'id'}, inplace = True)
```
